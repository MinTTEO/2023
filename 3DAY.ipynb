## 스케일링

## 클러스터링(군집화)
비지도 학습(라벨링X) : 정답 없이 데이터로 부터 중요한 의미를 찾아내는 머신러닝 기법, 성격이 비슷한 항목들끼리 그룹으로 묶는 작업  
**<span style="color:red"> 군집화에서 가장 중요한 작업으로는 적정한 굽집의 수(k)를 먼저 찾아야 함 </span>**  
**k(군집수)가 너무 작으면 여러가지 성격을 가진 항목들이 너무 섞여 있는 현상 발생**  
**k(군집수)가 너무 크면 항목들을 세밀하게 구분 할 수 있으나 그룹 수가 너무 많아져서 군집화를 하는 의미가 X**

### 유사도
항목간의 유사한 정도를 수치로 나타낸것(거리측정)  
메일이 스팸에 가까운지 아니면 정상 메일에 가까운지  
추천에서 두 아이템 또는 사람이 서로 얼마나 가까운지  
최적의 분석 결과가 나오도록 유사도를 변경해 가면서 반복  
유사도 s(simililarity)는 0 ≤ s ≤ 1(1에 가까울수록 유사도 높음)  
유사도의 상대 개념으로 거리(disranc)사용 : 유사도와 거리의 관계 d = 1 - s

## 클러스터링(KMeans) 수행
매우 쉬운 구현성, 높은 계산 효율성 => 하계와 산업현장을 가리지 않고 활약  
프로토타입 기반 군집 : 각 클러스터가 하나의 프로토타입으로 표현됨  
프로토타입(Prototype)  
연속적인 특성에서는 비슷한 데이터포인의 센트로이드(cetroid - 평균)  
범주현 특성에서는 메도이드(medoid - 가장 자주 등장하는 포인트)  
원형 클러스터 구분에 뛰어남, 사전에 몇개의 클러스터(군집수)를 만들 것인지 지정해줘야함 => 사람의 주관적인 판단 개입  
적절한 k값을 선택했다면 높은 성능이 나타날것이고, 아니면 부적합한 k값은 모델의 성능을 보장 할 수 없음 

##### KMeans Algorithm
1. Initialize k centroids.  k 중심을 초기화
2. Assign each data to the nearest centroid, these step will create clusters.   
각 데이터를 가장 가까운 중심에 할당하면 이 단계에서 클러스터가 생성
3. Recalculate centroid - which is mean of all data belonging to same cluster.  
동일한 클러스터에 속한 모든 데이터의 평균인 중심을 다시 계산
4. Repeat steps 2 & 3, till there is no data to reassign a different centroid.  
다른 중심을 재할당할 데이터가 없을때까지 2,3을 반복

```python
from sklearn.cluster import KMeans
k = KMeans(n_clusters = 4).fit(power1)
centers = k.cluster_centers_ # 4군집의 중심점(좌표) k개의 센트로이드를 기준으로 클러스터링을 진행하여 k개의 집단을 만든다.
print(centers)
```

## DBSCAN(Density based Spatial Clustering of Apllications with Noise)
### 밀도기반 클러스터링 알고리즘  
k-means처럼 단순히 거리만을 기준으로 군집화를 하는 것이 아니라  
**"가까이 있는 샘플들은 같은 군집에 속한다"** 는 원칙  
샘플들의 몰려있는 정도 즉, 밀도가 높은 부분을 중심으로 인접한 샘플들은 포함시켜 나간다.  
한 점을 기준으로 반경 r 내에 점이 n 개이상 있으면 하나의 군집으로 인식하는 방식

1번 데이터를 중심으로 보면 반지름 r인 원 안에 군집이 되기 위한 최소 기준인(예를들어 n=4(포인트(점)수 라면) 샘플이 없다.  
1번 데이터는 노이즈 데이터(noise point)가 되며 클러스터에 제외





3번과 8번 데이터를 중심으로 보면 원 안에 4개의 점이 있으며 이러한 데이터를 코어 데이터(core point)라고 한다.\
코어 데이터들은 스스로 클러스터를 형성 할 수 있다.




make_moons: 초승달 모양 클러스터 2개 형상의 데이터로 달이 두개가 서로 엎어져 겹친 모양  
이 모양을 서로 연결되어 있는 하나의 클러스터로 나누어져야 하는데  
k-means 알고리즘을 적용하면 거리가 가까운 점들을  
하나의 클러스터로 묶기 때문에 원하는 결과를 얻지 못함

scikit-learn 패키지는 분류(classification)모형의 테스트를 위해 여러가지 가상 데이터를 생성하는 함수를 제공  
make_classification함수는 설정에 따른 분류용 가상데이터를 생성하는 명령어  
make_blobs 함수를 등방성(방향과 상관없는 성질) 가우시안 정규분포를 이용해 가상 데이터를 생성  
make_moons 함수는 초승달 모양 클러스터 2개 형상의 데이터를 제공, make_moons 명령으로 만든 데이터는 직선을 사용하여 분류 할 수 없다. 
- n_features : 독립변수의 수, 디폴트 20 
- n_samples : 표본 데이터의 수, 디폴트 100
- noise : 잡음의 크기.0이면 정확한 반원을 이룸  
- centers : 생성할 클러스터의 중심(3개)
- cluster_std : 클러스터의 표준 편차 1

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_features = 2, n_samples = 1000,
                  centers = 3, cluster_std = 1, random_state = 3)

print(X)
plt.scatter(X[:,0], X[:,1], s = 5, alpha = .5)
```

```python
X, y = make_blobs(n_features = 2, n_samples = 1000, cluster_std = .5)

print(X)
plt.scatter(X[:,0], X[:,1], s = 5)
```

```python
kmeans = KMeans(n_clusters = 3)
predict = kmeans.fit_predict(X)
plt.scatter(X[:,0], X[:,1], s = 5, c = predict)
```

## Mean Shift
k-means와 유사하게 중심을 군집의 중심으로 지속적으로 움직이면서 군집화를 수행
### k-means가 중심에 소속된 데이터의 평균 거리 중심으로 이동하는데 반해,
### Mean Shift는 중심을 데이터가 모여있는 밀도가 가장 높은 곳으로 이동


### Mean Shift는 데이터의 분포도를 이용해 군집 중심점을 찾는다  
### 군집 중심점은 데이터 폰인트가 모여 있는 곳이라는 생각에서 착안한 것이며,
### 이를 위해 확률밀도점수(P.D.F)를 이용한다.

```python
from sklearn.cluster import KMeans, MeanShift
ms = MeanShift()
predict = ms.fit_predict(X)
plt.scatter(X[:,0], X[:,1], s = 5, c = predict)
```
